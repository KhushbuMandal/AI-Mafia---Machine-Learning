# Example 5

**Plot histogram** 

```python
ages = [2, 45, 67 , 3 , 67, 89, 3 , 30 , 45, 32 ,7 , 13, 57, 18 , 90 , 77, 21 , 20 ,40]
```

```python
#setting the ranges and no. of intervals
range = (0 ,100)
bins = 10
```

```python
#plotting a histogram
plt.hist(ages , bins , range, color = 'red',
        histtype = 'bar', rwidth= 0.8)
plt.xlabel('age') # to label the graph at x axis 
plt.ylabel('no of people') #to label the graph at y axis 
plt.title('My HIstogram')
```

```python
plt.show()
plt.close()
```

# Example 7

**Plotting Pie-Chart**

```python
#defining labels
activities = ['eat', 'sleep', 'work', 'play']
```

```python
#portion covered by each label
slices = [3, 7, 8, 6]
```

```python
#color for each label
colors = ['r', 'm', 'g', 'b']
```

```python
#plotting the pie chart
plt.pie(slices, labels = activities, colors = colors,
       startangle = 90, shadow = True, explode = (0, 0, 0.1, 0),
       radius = 1.2, autopct = '%f %%')
plt.title('My Pie Chart')
```

```python
plt.show()
plt.close()
```

# Example 9

**Plotting subplots using method 1**

```python
#function to generate coordinates
def create_plot(ptype):
    #setting the x-axis values
    x = np.arange(-10, 10, 0.01)
    
    #setting the Y-axis values
    if ptype == 'linear' :
        y = x
    elif ptype == 'quadratic':
        y = x**2
    elif ptype == 'cubic':
        y = x**3
    elif ptype == 'quartic':
        y = x**4 
        
    return (x, y)    
        
```
